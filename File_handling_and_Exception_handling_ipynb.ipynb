Pratice problem-01

```python
from functools import reduce

with open ("./sample_data/sample.txt") as file :
  strings = file.readlines()

  total_lines = len(strings)
  print(total_lines)

  #word count ber korbo
  word_count = list(map(lambda x : len(x.split()),strings))
  print(word_count)

  total_words = reduce(lambda x,y : x+y,word_count)
  print(total_words)

  #cleaning process

  #1.New line deleting
  strings = list(map(str.strip,strings))
  print(strings)

  #2.speace delete

  strings = list(map(lambda x : x.replace (" ","") ,strings))
  print(strings)

  #3.count of character

  count_character = list(map(lambda x : len(x),strings))
  print(count_character)
  total_character = reduce(lambda x,y : x+y,count_character)
  print(total_character)


  # Now store the values in new txt file


  with open("./sample_data/string.txt","w") as file :
    file.write(f"total lines:{total_lines}\n number of words:{total_words}\n number of total character:{total_character}")

```

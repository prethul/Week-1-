Different type of argument

```python
def adder (a,b,c):
  print(f"{a},{b},{c}")
adder(10,20,30)
```

```python
def adder (*args):
  for i in args:
    print(i,end=" ")
adder(10,20,30,45,76,39,67)
```

```python
def adder(a,b,c):

  print(f"{a},{b},{c}")

adder(a=10,c=20,b=30)
```

```python
def adder(**kwargs):
  print(kwargs)
adder(a=10,b=30,c=20,n_estimator = 50)
```

```python
def adder(**kwargs):
  for key,value in kwargs.items():
    print(f"{key} : {value}")
adder(a=10,b=30,c=20,n_estimator = 50)
```

```python
#pratice the same code

def adder(**kwargs):
  for key,value in kwargs.items():
    print(f"{key} : {value}")
adder(a=10, b = 30, c = 40 , n_estimator = 60)
```

Return Type

```python
def welcome_message():
  print("Hello world")
print(type(welcome_message()))
```

```python
def adder(a,b):
  sum = a+b
  return sum
print(type(adder(10,20)))
```

```python
def welcome_message(user ="none"):
  return f"Hello {user}"
print(type(welcome_message()))
```

```python
def adder (a,b):
  return a,b
print(type(adder(10.5,20)))

x,y = adder(10.5,20)
print(x,y)
```

Generator


```python
lst =[x for x in range(500)]

def data_loader(chunk_size,lst):
  for i in range(0,len(lst),chunk_size):
    yield lst [i:i+chunk_size]
x = data_loader(5,lst)

print(next(x))
print(next(x))
```
